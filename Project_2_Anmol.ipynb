```python
# create Dash app
external_stylesheets = [dbc.themes.BOOTSTRAP]
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.title = 'Supermart Grocery Sales Dashboard'
```

```python
df2= data.groupby(['Region', 'Sub Category'])[['Sales', 'Profit']].sum()
df2["Region"] = df2.index
df2
```

```python
df1= data.groupby(['Region', 'Category'])[['Sales', 'Profit']].sum()
df1["Region"] = df1.index
df1
```

```python
# define the layout
regions = data["Region"].sort_values().unique()
app.layout = html.Div(
    children=[
        html.H1(children="Supermart Grocery Sales Dashboard",className="header-title",
                style={"fontSize": "48px", "color": "blue", "textAlign":"center"}),

        html.Div(
    children=[
        html.Div(children="Region", className="menu-title"),
        dcc.Dropdown(
            id="region-filter",
            options=[
                {"label":region, "value":region}
                for region in regions
            ],
            value="South",
            clearable=False,
            className="dropdown",
        ),
    ]
),


        dcc.Graph(id='graph1'),
        dcc.Graph(id='graph2'),
        dcc.Graph(id='graph3'),
        dcc.Graph(id='graph4')

    ]
)

@app.callback(
    Output("graph1", "figure"),
    [Input("region-filter", "value")]
)
def update_graph1(region):
    dff = data[data["Region"] == region]
    df2 = dff.groupby("Category").sum().reset_index()
    return {"data": [{"x": df2["Category"], "y": df2["Sales"], "type": "lines"}],
            "layout": {"title": f"Categorywise Total Sales in {region}"}}


@app.callback(
    Output("graph2", "figure"),
    [Input("region-filter", "value")]
)
def update_graph2(region):
    dff = data[data["Region"] == region]
    df2 = dff.groupby("Category").sum().reset_index()
    return {"data": [{"x": df2["Category"], "y": df2["Profit"], "type": "bar"}],
            "layout": {"title": f"Categorywise Total Profit in {region}"}}


@app.callback(
    Output("graph3", "figure"),
    [Input("region-filter", "value")]
)
def update_graph3(region):
    dff = data[data["Region"] == region]
    df1 = dff.groupby("Sub Category").sum().reset_index()
    return {"data": [{"x": df1["Sub Category"], "y": df1["Sales"], "type": "lines"}],
            "layout": {"title": f"Sub-Categorywise Total Sales in {region}"}}


@app.callback(
    Output("graph4", "figure"),
    [Input("region-filter", "value")]
)
def update_graph4(region):
    dff = data[data["Region"] == region]
    df1 = dff.groupby("Sub Category").sum().reset_index()
    return {"data": [{"x": df1["Sub Category"], "y": df1["Profit"], "type": "bar"}],
            "layout": {"title": f"Sub-Categorywise Total Profit in {region}"}}
```

```python
if __name__ == "__main__":
    app.run_server(debug=False, port = 8070)
```
